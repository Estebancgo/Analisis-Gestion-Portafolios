Proyecto: Estimación de retiros Netos para administración de recursos de terceros
==

1.Objetivo del negocio
==
> En una economía de libre mercado, cuando los diferentes agentes que en esta interactúan acumulan excesos de liquidez, se producen oportunidades de negocio para las instituciones financieras, las cuales se encuentran en facultad de ofrecer productos de inversión en el mercado de valores a todo el público, con el fin de obtener un beneficio adicional para sus clientes a cambio del cobro de una comisión por sus servicios.

> Naturalmente dentro del proceso de vida de un ser humano, se busca lograr estabilidad económica al momento de llegar a la edad pensional. En ese sentido un vehículo a través del cual una persona complementa su proceso pensional son los Fondos de Pensiones Voluntarias, los cuales permiten acumular un ahorro que las entidades financieras invierten en el mercado de valores, de tal forma que llegado el momento se puedan complementar los aportes obligatorios para así tener una mesada pensional más alta. Una entidad que administra este tipo de productos está interesada en los siguientes propósitos dentro de los servicios que comercializa:

>- Ofrecer un producto que respete el mandato del cliente.
>- Presentar una oferta competitiva dentro del mercado.
>- Entregar resultados que beneficien a sus clientes acorde a los movimientos del mercado de valores.
>- Administrar de forma adecuada los riesgos inmersos en los procesos de inversiones.
>- Respetar los lineamientos legales que establezcan los entes de control.

> Fiduciaria Davivienda, es una entidad que tiene como propósito ofrecer productos Fiduciarios al público en general, entre los cuales destacan los Fondos de Inversión Colectiva (FIC), los Fondos Voluntarios de Pensiones (FVP) y los diferentes servicios de Fiducia estructurada como lo es por ejemplo la fiducia de pagos. Para la gestión de este tipo de productos interactúan diversas áreas que permiten realizar una adecuada administración de los recursos entregados por sus clientes, entre estas se encuentran:

>- **Inversiones**: Es un área encargada de canalizar los recursos provenientes de los clientes en el mercado de valores, de tal forma que se busque maximizar el beneficio obtenido para los mismos, esto se realiza a través de un portafolio de inversiones, este último es un compendio de activos financieros que emiten empresas tanto locales como internacionales (mercado primario) o que luego de su emisión son comprados y vendidos por los agentes del mercado de valores (mercado secundario)
>- **Riesgos**: Es el pilar que permite discernir no solo el cumplimiento de los límites en el proceso de inversiones, sino que aporta herramientas que buscan apoyar la adecuada gestión del equipo de inversiones, para ofrecer un producto de calidad.
>- **Comercial**: Es un equipo encargado de gestionar el posicionamiento de los productos de la organización en el mercado, de tal forma que se incrementen el número de clientes que hacen parte de las alternativas de inversión ofrecidas.




> Descripción del problema
--

> Como se mencionó previamente parte de la gestión del equipo de inversiones está enmarcada en trasladar los recursos aportados por sus clientes al mercado de valores, sin embargo este proceso contiene ciertos riesgos inherentes, que pueden afectar su objetivo primario, entre estos se encuentra el riesgo de liquidez, el cual se materializa en el evento en el que los requerimientos de los clientes para retirar su dinero, sea mayor a los recursos que se tengan en cuentas de ahorro en el portafolio. Si bien, en el escenario en el que la organización no cuente con los recursos disponibles (por estar invertidos en el mercado de valores) no implica que el cliente pierde su dinero, si podría ocasionar una serie de consecuencias reputacionales y legales para la organización.

> Como parte de un proceso para reducir la probabilidad de que se presente una situación como la mencionada, las entidades, para los FVP a manera de buena práctica, construyen modelos que intentan medir los niveles de liquidez y establecer cuantas veces está en capacidad de responder ante un retiro probable de acuerdo a movimientos históricos de dinero por parte de los clientes.

> Situación actual
--

> Actualmente En el caso de los FVP la normativa vigente no establece ningún modelo especial que se deba calcular para mitigar el riesgo de liquidez mencionado previamente, es por ello que recae en la entidad la responsabilidad de contar con un modelo que le permita estimar los recursos con los que cuenta para responder por las necesidades de retiros de los clientes. Para ello se puede estimar un retiro probable a partir de la historia, usando el calculo de un percentil del nivel de retiro de los clientes, cuanta más historia se tenga mayor cantidad de eventos del pasado se pueden incluir para entender el comportamiento del futuro.

> A continuación, se enuncian algunas ventajas de este tipo de soluciones:

> - Fácil implementación computacional, inclusive no requiere necesariamente de ninguna clase de programación
> - Es una metodología sencilla de entender en general para cualquier equipo de personas que intervengan en el proceso de inversiones
> - Consume un muy corto tiempo de ejecución 

> Sin embargo, también se pueden ver algunas desventajas:

> - Estos métodos suelen responder conforme se presentan las situaciones, es decir, no cuentan con un gran poder predictivo, e inclusive levantan alertas cuando ya suele ser tarde  para gestionar los movimientos de los clientes.

> - Se tiene poca sensibilidad ante choques leves en términos de retiros de los clientes, los cuales también pueden afectar de forma especial la gestión desde el ámbito de inversiones.

> - Al ser muchas veces métricas que sólo generan alertas en momentos de altos niveles de retiros, suele pasar que no son usadas con regularidad, porque en la mayoría de días muestran que se cuentan con niveles adecuados de recursos disponibles para atender los retiros de los clientes; es en este punto donde se materializa el éxito del presente proyecto, ya que al tener una métrica más exacta siendo menos ácida se logrará un uso más frecuente y sobre todo un uso que apoye la toma de decisiones del equipo de inversiones en el día a día.

> El negocio por su parte busca dinamizar el seguimiento sobre los niveles de retiros de sus clientes, esto permitirá mejorar el proceso de inversiones ya que naturalmente desde una perspectiva de rentabilidad lo más conveniente no es mantener la mayor parte de los recursos en cuentas de ahorro.


2.Objetivo de la minería de Datos
==

> El presente proyecto busca establecer una técnica de analítica predictiva para series de tiempo, que permita calcular de forma diaria el nivel de retiro neto probable a 1 día de los diferentes portafolios que componen el Fondo Voluntario de Pensiones de la organización. En torno a esta definición se desprenden algunos propósitos específicos que enmarcan el proyecto:

> - Desarrollar un modelo de fácil implementación computacional, el cual busque la menor interacción por parte del área encargada, de tal forma que su ejecución sea lo más practica y automática posible.

> - Propender por usar la información disponible públicamente, de la forma en que se comportan los retiros en los fondos de la industria que ofrecen un servicio similar.

> - Explicar de forma sencilla la metodología con la que funciona el modelo, así como las definiciones de los supuestos y características que se usaran en el mismo.

> Riesgos del proyecto
>---

> - El primer riesgo se encuentra asociado al tiempo, ya que dada la duración máxima de aproximadamente 2 meses, podría darse el escenario en el cual el mismo no sea suficiente para lograr la totalidad de los desarrollos y evaluaciones necesarias

> - De igual forma la cantidad de variables directamente relacionadas con el proceso de retiros por parte de los clientes son limitadas, lo cual puede restar precisión al modelo

> - Cabe la posibilidad de que a pesar de ser una técnica predictiva, su capacidad no sea lo suficientemente buena y que por tal motivo resulte siendo más práctico continuar con los cálculos actualmente existentes

> - A pesar de que el modelo entregue resultados con un nivel muy alto de aproximación a la realidad, se puede presentar resistencia a adoptar el mismo, y esto se da principalmente por la incertidumbre que entrega el hecho de que se deje total confianza a la estimación de un modelo y que en cierto escenario pueda mostrar un valor errado que genere problemas para la organización

> - La disponibilidad de los datos públicos de la competencia puede tener un nivel de rezago importante, y como los datos requieren ser usados diariamente, es posible que se deba trabajar únicamente con los datos propios de la entidad

>El proyecto no requiere de la inversión de recursos directamente al tener información disponible pública e internamente, esto lo hace una iniciativa totalmente viable para la organización, además que por sí sola plantea un escenario de disrupción a las técnicas clásicas usadas dentro de los proceso de inversiones.

> En ese sentido la minería apoya al negocio y está alineada con los objetivos de este, ya que sus resultados permitirán una gestión de la liquidez de los portafolios administrados, más adecuada a la realidad y que impacte positivamente los resultados que obtiene la organización para sus clientes

3.Criterios de Éxito
==

> - El modelo debe poder pronosticar con la mayor precisión los niveles de retiros netos, a través de un criterio de negocio se establece que debe estar dentro de un rango del 5% de desviación para que realmente sea usado en el día a día del negocio.

> - Se espera que la ejecución diaria no implique importantes traumatismos para el área encargada, ya que de lo contrario no se convertiría realmente en una solución adecuada.

> - La forma en la que se explique la construcción del modelo se debe realizar de foma sencilla para que así mismo sea adoptado de forma positiva por el área de inversiones

> - Se debe tener la capacidad de reajustar el modelo en el escenario en que su precisión disminuya, sin que esto implique una total reformulación.
